```python
def assignToCustomerGroup(x):
  result = "Unknown"
  if x in range(20000):
    result = "Very Low Income"
  elif x in range(20000, 40000):
    result = "Low Income"
  elif x in range(40000, 60000):
    result = "Medium Income"
  elif x in range(60000, 80000):
    result = "High Income"
  elif x in range(80000, 200000):
    result = "Very High Income"
  elif x in range(100000, 200000):
    result = "Extremely High Income"
  return result
```

```python
parameter = dfSample["Country"].isin(["Canada", "France", "Germany"])
dfSampleCountries = dfSample[parameter]
dfSampleCountries["Customer Category"] = dfSampleCountries["Annual Income"].apply(assignToCustomerGroup)
```
